```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gallery(array, ncols=3):
    nindex, height, width, intensity = array.shape
    nrows = nindex//ncols
    assert nindex == nrows*ncols
    # want result.shape = (height*nrows, width*ncols, intensity)
    result = (array.reshape(nrows, ncols, height, width, intensity)
              .swapaxes(1,2)
              .reshape(height*nrows, width*ncols, intensity))
    return result
```

```python
def make_array():
    from PIL import Image
    return np.array([np.asarray(Image.open('casita.png').convert('RGB'))]*12)
```

```python
array = make_array()
array[0]
```

```python
result = gallery(array)
```

```python
import pandas as pd
data = pd.read_csv('data_earthquake_2000-2017.csv',encoding='latin1', sep=';')
```

```python
    old_max = data['Magn'].max()
    old_min = data['Magn'].min()
    old_range =  old_max - old_min
    new_max = 255
    new_min = 40
    new_range = new_max - new_min 
    data['Magn'] = data['Magn'].apply(lambda v: int(((v-old_min)*new_range)/old_range)+new_min)
```

```python

from mpl_toolkits.basemap import Basemap   
```

```python
print(data.shape)
data = data[data['Prof.'] >= 60.0 ]
print(data.shape)
```

```python
lat = data['Latitud'].values
lon = data['Longitud'].values
magn = data['Magn'].values
prof = data['Prof.'].values
```

```python
lat_max = data['Latitud'].max()
lat_min = data['Latitud'].min()
lon_max = data['Longitud'].max()
lon_min = data['Longitud'].min()
lat_m = data['Latitud'].mean()
lon_m = data['Longitud'].mean()
```

```python
fig = plt.figure(figsize=(6, 6))
m = Basemap(llcrnrlon=lon_min,llcrnrlat=lat_min,urcrnrlon=lon_max,urcrnrlat=lat_max,
            resolution='i',projection='lcc',lat_0=lat_m,lon_0=lon_m)
m.shadedrelief()
m.drawcoastlines(color='gray')
m.drawcountries(color='gray')

m.scatter(lon, lat, latlon=True,
          c=magn, 
          cmap='Reds', alpha=0.5)

plt.show()
plt.savefig('tumama.jpg')
```

```python
# 1. Draw the map background
#-36.739055 -71.0574941
fig = plt.figure(figsize=(8, 8))
m = Basemap(projection='lcc', resolution='h', 
            lat_0=-35, lon_0=-70,
            width=1E6, height=2.3E6)
m.shadedrelief()
m.drawcoastlines(color='gray')
m.drawcountries(color='gray')


# 2. scatter city data, with color reflecting population
# and size reflecting area
m.scatter(lon, lat, latlon=True,
          c=np.log10(prof), s=magn,
          cmap='Reds', alpha=0.5)


# make legend with dummy points
```
